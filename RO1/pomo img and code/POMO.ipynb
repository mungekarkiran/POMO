```python
import glob, shutil, random, os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image, ImageEnhance
from sklearn.model_selection import train_test_split
from tqdm import tqdm
```

```python
# Make .csv with height and weight

folder,count = [],[] 

for i in glob.glob(r'img\*'):
#     print(i)
#     print(len(glob.glob(i+'\*')))
    folder.append(i)
    count.append(len(glob.glob(i+'\*')))
    
df = pd.DataFrame(columns=['class', 'count'])
df['class'] = folder
df['count'] = count
df.to_csv(r'class_count.csv', index=False)
df
```

```python

img_path, img_format, mode, we, he = [],[],[],[],[]

for i in glob.glob(r'img\*'):
    print(i)
#     print(glob.glob(i+'\*'))
    for img in glob.glob(i+'\*'):
#         print(img)
        img_path.append(img)
        image = Image.open(img)
#         To show images.
#         image.show()
#         The file format of the source file.
#         print(image.format)
        img_format.append(image.format)
#         The pixel format used by the image. Typical values are "1", "L", "RGB", or "CMYK."
#         print(image.mode)
        mode.append(image.mode)
#         Image size, in pixels. The size is given as a 2-tuple (width, height).
        w, h = image.size 
#         print(w, h)
        we.append(w)
        he.append(h)

df = pd.DataFrame(columns=['img_path', 'img_format', 'img_mode', 'img_width', 'img_height'])
df['img_path'] = img_path
df['img_format'] = img_format
df['img_mode'] = mode
df['img_width'] = we
df['img_height'] = he

df.to_csv(r'data_information.csv', index=False)
df
```

```python
# resize image
c=0

for i in glob.glob(r'img\*'):
    print(i)
    for img in glob.glob(i+'\*'):
        print(img)
        image = Image.open(img)
        w, h = image.size 
        print(w,h)
#         Resizing Images
        if (w < 224 or h < 224):
            new_image = image.resize((224, 224))
            print(new_image.size)
            new_image.save(i+f'\Resized_img_{c}.jpg')
            c+=1
```

```python
# movie images
c=0
for i in glob.glob(r'img\*'):
    print(i)
    for img in glob.glob(i+'\*'):
        print(img)
        image = Image.open(img)
        w, h = image.size 
        print(w,h)
#         moving Images
        if (w < 224 or h < 224): pass
        else: 
            dst = i[-1]
            print(img, f'img1/{dst}/{dst}_{c}.jpg')
            shutil.copy2(img, f'img1/{dst}/{dst}_{c}.jpg')
            c+=1
```

```python
# Rotating Images
c=0
right = 100//2
left = 100//2
top = 100//2
bottom = 100//2

for i in glob.glob(r'img1\*'):
    print(i)
    for img in glob.glob(i+'\*'):
#         print(img)
        image = Image.open(img)
        ang = random.randint(20, 30)
        image_rot = image.rotate(ang, expand=True)
        image_rot.save(i+f'\Rotated1_img_{c}.jpg')
#         image_flip = image.transpose(Image.FLIP_LEFT_RIGHT)
#         image_flip.save(i+f'\Fliped1_img_{c}.jpg')
        contrast = ImageEnhance.Contrast(image)
        contrast.enhance(1.8).save(i+f'\Contrasted1_img_{c}.jpg')
        color = ImageEnhance.Color(image)
        color.enhance(1.7).save(i+f'\Colored1_img_{c}.jpg')
        brightness = ImageEnhance.Brightness(image)
        brightness.enhance(1.4).save(i+f'\Brightness1_img_{c}.jpg')
        sharpness = ImageEnhance.Sharpness(image)
        sharpness.enhance(1.4).save(i+f'\Sharpness1_img_{c}.jpg')
#         To padding
        width, height = image.size
        new_width = width + right + left
        new_height = height + top + bottom
        padding = Image.new(image.mode, (new_width, new_height), (255, 255, 255))
        padding.paste(image, (left, top))
        padding.save(i+f'\Padding_img_{c}.jpg') 
        
        c+=1
```

```python
img_path, img_format, mode, we, he, y_class = [],[],[],[],[],[]

for i in glob.glob(r'img\*'):
    print(i)
#     print(glob.glob(i+'\*'))
    for img in glob.glob(i+'\*'):
#         print(img)
#         print(img.split("\\")[1])
        img_path.append(img)
        image = Image.open(img)
#         To show images.
#         image.show()
#         The file format of the source file.
#         print(image.format)
        img_format.append(image.format)
#         The pixel format used by the image. Typical values are "1", "L", "RGB", or "CMYK."
#         print(image.mode)
        mode.append(image.mode)
#         Image size, in pixels. The size is given as a 2-tuple (width, height).
        w, h = image.size 
#         print(w, h)
        we.append(w)
        he.append(h)
#         Actual class 
        y_class.append(img.split("\\")[1])
        
df = pd.DataFrame(columns=['img_path', 'img_format', 'img_mode', 'img_width', 'img_height', 'class'])
df['img_path'] = img_path
df['img_format'] = img_format
df['img_mode'] = mode
df['img_width'] = we
df['img_height'] = he
df['class'] = y_class

df.to_csv(r'data_information.csv', index=False)
df
```

```python
X, Y = df.iloc[:,:-1], df.iloc[:,-1]
# X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state = 42, stratify = Y)
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state = 101, stratify = Y)
```

```python
# data set of 80-20%

df1 = pd.DataFrame(X_train, columns=['img_path', 'img_format', 'img_mode', 'img_width', 'img_height', 'class'])
df1['class'] = y_train
df1 = df1.reset_index(drop=True)
# display(df1)

df2 = pd.DataFrame(X_test, columns=['img_path', 'img_format', 'img_mode', 'img_width', 'img_height', 'class'])
df2['class'] = y_test
df2 = df2.reset_index(drop=True)
# display(df2)

if os.path.isdir(f"csv_data"):
    df1.to_csv(f'csv_data/train2_82p.csv', index=False)
    df2.to_csv(f'csv_data/test2_82p.csv', index=False)
else:
    os.makedirs(f'csv_data')
    df1.to_csv(f'csv_data/train2_82p.csv', index=False)
    df2.to_csv(f'csv_data/test2_82p.csv', index=False)
```
