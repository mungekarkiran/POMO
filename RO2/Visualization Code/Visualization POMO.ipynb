```python
file1 = open("vgg16.txt","r")
line = []
for ind_line in file1.readlines():
    print(ind_line)
    line.append(ind_line)
    
```

```python
epoch = []
for ind_line in range(len(line)):
    if ind_line%2 == 0:
        print(ind_line, line[ind_line].split(' ')[1].split('/')[0])
        epoch.append(line[ind_line].split(' ')[1].split('/')[0])
```

```python
loss, accuracy, val_loss, val_accuracy = [],[],[],[] 
for ind_line in range(len(line)):
    if ind_line%2 == 1:
#         print(ind_line, line[ind_line].split(' '))
#         print(ind_line, line[ind_line].split(' ')[7])
#         print(ind_line, line[ind_line].split(' ')[10])
#         print(ind_line, line[ind_line].split(' ')[13])
#         print(ind_line, line[ind_line].split(' ')[16].split('\n')[0])
        loss.append(line[ind_line].split(' ')[7])
        accuracy.append(line[ind_line].split(' ')[10])
        val_loss.append(line[ind_line].split(' ')[13])
        val_accuracy.append(line[ind_line].split(' ')[16].split('\n')[0])
        
```

```python
vgg16_df = pd.DataFrame()
vgg16_df['epoch'] = epoch
vgg16_df['loss'] = loss
vgg16_df['accuracy'] = accuracy
vgg16_df['val_loss'] = val_loss
vgg16_df['val_accuracy'] = val_accuracy

vgg16_df.to_csv('vgg16_training.csv', index=False)
```

```python
file1 = open("Xception.txt","r")
line = []
for ind_line in file1.readlines():
    line.append(ind_line)
    
epoch = []
for ind_line in range(len(line)):
    if ind_line%2 == 0:
        epoch.append(line[ind_line].split(' ')[1].split('/')[0])

loss, accuracy, val_loss, val_accuracy = [],[],[],[] 
for ind_line in range(len(line)):
    if ind_line%2 == 1:
        loss.append(line[ind_line].split(' ')[7])
        accuracy.append(line[ind_line].split(' ')[10])
        val_loss.append(line[ind_line].split(' ')[13])
        val_accuracy.append(line[ind_line].split(' ')[16].split('\n')[0])

vgg16_df = pd.DataFrame()
vgg16_df['epoch'] = epoch
vgg16_df['loss'] = loss
vgg16_df['accuracy'] = accuracy
vgg16_df['val_loss'] = val_loss
vgg16_df['val_accuracy'] = val_accuracy

vgg16_df.to_csv('Xception_training.csv', index=False)
```
